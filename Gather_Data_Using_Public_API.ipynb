```python
import requests #digunakan untuk request data dari API
import json #digunakan untuk merubah format dalam json agar hasil data dapat dibaca dengan mudah
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm #digunakan untuk menambahkan progress bar pada perulangan
```

```python
#fungsi untuk mengambil data

def searchloc(location):
  req_string = f"https://www.onemap.gov.sg/api/common/elastic/search?searchVal={location}&returnGeom=Y&getAddrDetails=Y&pageNum=1" #berisi url API
  resp = requests.get(req_string) #perintah meminta request API
  resp = json.loads(resp.content) #merubah format menjadi json agar tampilan hasil mudah dibaca (dictionary)
  return resp['results'][0] #mengembalikan hasil 'results' pada indeks pertama, karena 1 nama lokasi dapat muncul lebih dari 1 lokasi
```

```python
#list untuk menyimpan data lokasi MRT
result_list = []

#perulangan untuk mencari data lokasi MRT pada setiap iterasi dari list MRT yang akan dicari
for i in tqdm(mrt_stations):
  result_list.append(searchloc(i))
```
